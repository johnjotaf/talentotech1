```python
import pandas as pd

# Create a dataset
# Let´s build the values lists
value_n_list = ["Andrew","Phillip","John","Jose","Mario","Carlos","Rafael","Luis","Diego","Alex"]
print(value_n_list)

value_w_list = [50,    78,    91.5,   79, 80, 100, 55, 66, 88.5, 45] # [kg]
value_h_list = [1.76,   1.87,   1.69, 1.80, 1.65, 1.97, 1.50, 1.66, 1.55, 1.98] # [m]
# value_bmi_list will be calculated: bmi = w/(h^2)
#value_bmi_list = [25,25,25] # init values [kg/m2]

# Let´s use a disctionary ({key:value})
# with the format {"str":list}
data_dict = {"Name":value_n_list,
             "Weight":value_w_list,
             "Height":value_h_list} #,
             #"BMI":value_bmi_list}

print(data_dict)


# Turn a dictionary in to a dataframe
df = pd.DataFrame(data_dict)
print(df)
#call a variable
df

df["BMI"]=df["Weight"]/df["Height"]**2
df
```

```python
# prompt: Mediante el marco de datos df: hazme una linea de codigo para utilizar el anterior dataframe y clasificar segun el BMI de la tabla

# Define a function to classify BMI
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal weight'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obesity'

# Apply the classification function to the 'BMI' column
df['BMI_Category'] = df['BMI'].apply(classify_bmi)
```

```python
# Calculadora cientifica con funcion While
def suma(a, b):
    return a + b
def multiplicar(a, b):
    return a * b
def divicion(a, b):
    return a / b
def resta(a - b):
    return a - b
while True:
    print("Menu Principal")
    print("1. Sumar")
    print("2. multiplicar")
    print("3. dividir")
    print("4. resta")
    opc = input("Ingrese una opcion: ")
    n1 = float(input("ingrese el primer numero;"))
    n2 = float(input("ingrese el segundo numero"))
    if opc == "1":
        print("La suma es:", suma(n1, n2))
    elif opc == "2"
        print("La multiplicacion es:", multiplicar(n1, n2))
    elif opc == "3":
        print("La divicion es:", divicion(n1, n2))
    elif opc == "4":
        print("La resta es:", resta(n1, n2))
```
