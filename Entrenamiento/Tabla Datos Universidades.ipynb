```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Datos de ejemplo
# Supongamos que estos son datos reales de, por ejemplo, horas de estudio vs. puntaje de examen
horas_estudio = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10]).reshape(-1, 1) # Característica (variable independiente)
puntaje_examen = np.array([50, 60, 65, 70, 75, 80, 85, 90, 95]) # Variable objetivo (variable dependiente)

# Crear el modelo de regresión lineal
modelo_regresion = LinearRegression()

# Entrenar el modelo
modelo_regresion.fit(horas_estudio, puntaje_examen)

# Realizar predicciones
puntaje_predicho = modelo_regresion.predict(horas_estudio)

# Visualizar los resultados
plt.scatter(horas_estudio, puntaje_examen, color='blue')
plt.plot(horas_estudio, puntaje_predicho, color='red')
plt.title('Regresión Lineal Simple')
plt.xlabel('Horas de Estudio')
plt.ylabel('Puntaje de Examen')
plt.show()

# Imprimir los coeficientes del modelo
print("Intercepción (b0):", modelo_regresion.intercept_)
print("Coeficiente (b1):", modelo_regresion.coef_[0])
```
